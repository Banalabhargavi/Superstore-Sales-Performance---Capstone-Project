```python
# Importing the required libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# Load the dataset
df =pd.read_excel('Dataset_1.xlsx')
```

```python
#Display the first few rows
df.head()
```

```python
# Select relevant columns and drop missing values
df_clean = df[['PRICE', 'PRICE_SQFT', 'NO_OF_BATHROOMS', 'FLOOR_SIZE']].dropna()
```

```python
# Define independent variables (price per square foot, number of bathrooms, floor size)
X = df_clean[['PRICE_SQFT', 'NO_OF_BATHROOMS', 'FLOOR_SIZE']]
X = sm.add_constant(X)  # Add constant term for intercept

# Define dependent variable (Price)
Y = df_clean['PRICE']
```

```python
# Fit the multiple linear regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print summary of the model
print(results.summary())
```

```python
# Compute 95% confidence intervals for regression coefficients
confidence_intervals = results.conf_int(alpha=0.05)
confidence_intervals.columns = ['Lower Bound', 'Upper Bound']
print(confidence_intervals)
```
